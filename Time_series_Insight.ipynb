```python
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load the dataset
df = pd.read_csv("https://gist.githubusercontent.com/denandreychuk/b9aa812f10e4b60368cff69c6384a210/raw/100%20Sales%20Records.csv")

#check the data format
print(df["Order Date"].head())
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Order Month"] = df["Order Date"].dt.to_period("M")
monthly_revenue = df.groupby("Order Month")["Total Revenue"].sum()
print(monthly_revenue)

plt.figure(figsize=(12,6))
monthly_revenue.plot(kind='line', marker='o',color='teal')

plt.title("Total Revenue Over Time", fontsize=16)
plt.xlabel("Order Month")
plt.ylabel("Total Revenue")
plt.grid(True)
plt.tight_layout()
plt.show()
```
